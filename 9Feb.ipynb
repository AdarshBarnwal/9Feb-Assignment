```python
Q3.
Ans:-
      Multiple inheritance is a feature of object-oriented programming languages where a class can inherit properties
     and methods from multiple parent classes.
        
# Example

class father:
    def Arrav_father(self):
        print("my name in divyansh and my son name is Arrav.")
        
class mother:
    def Arrav_mother(self):
        print("my name is hunny and my son name is Arrav.")
        
class child(father,mother):
    def Arrav(self):
        print("My name is Arrav and my fathers name is divyansh and my mother name is hunny")
        
obj_child=child

obj_child.Arrav_father()
obj_child.Arrav_mother()
obj_child.Arrav
```

```python
Q4.
Ans:-

     getters and setters are methods used to access and modify the values of instance variables (or attributes) of a class.
    Getters are used to access the value of an attribute, while setters are used to set or modify the value of an attribute
    
class check_getter_and_setter:                #creating a class
    
    def __init__(self,book_name,price):        #assingment value
        
        self.book_name=book_name
        self.__price=price
        
    @property
    def price_access(self):                 #getter i.e allow to give access aprivate attribute
        
        return self.__price
    
    @price_access.setter                       #setter i.e allow to set or mutate the value
    def price_update(self,new_price): 
        self.__price=new_price
        return self.__price 
    
obj=check_getter_and_setter("let us python",280)          #creating instance or object
print(obj.book_name)
obj.price_access         # it will give price 280
obj.price_update=265
obj.price_access    # it will give price 265
      
     
```
